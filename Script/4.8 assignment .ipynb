```python
# 6 identify the spender 
```

```python
# create average_price per user spend 
ords_prods_merge['avg_price']=ords_prods_merge.groupby(['customer_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
#create spending_flag
ords_prods_merge.loc[ords_prods_merge['avg_price'] <10, 'Spending_flag'] = 'Low Spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] >=10, 'Spending_flag'] = 'High Spender'
```

```python
ords_prods_merge['Spending_flag'].value_counts(dropna=False)
```

```python
#7 frequent user vs. non-frequent user 
```

```python
#create median order duration 
ords_prods_merge['median_order_duration']=ords_prods_merge.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
```

```python
#create frequency_flag
ords_prods_merge.loc[ords_prods_merge['median_order_duration'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_order_duration'] > 10) & (ords_prods_merge['median_order_duration'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_order_duration'] < 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
